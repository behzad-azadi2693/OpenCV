### read camera and save that

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('out.avi', fourcc, 24.0, (640, 480))

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    out.write(frame)

    cv2.imshow('Webcam', gray)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

### add text to image

### threshold image

```python
img  = cv2.imread('image.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
th = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 115, 1)

cv2.imshow('test', th)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### mask camera

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower = np.array([100,0,0])
    upper = np.array([255,255,255])
    mask = cv2.inRange(hsv, lower, upper)
    res = cv2.bitwise_and(frame, frame, mask=mask)


    cv2.imshow('webcam', res)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

### the edge of the image

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    laplacian = cv2.Laplacian(frame, cv2.CV_8U)
    sobelx = cv2.Sobel(frame, cv2.CV_8U, 1, 0, ksize=5)
    sobely = cv2.Sobel(frame, cv2.CV_8U, 0, 1, ksize=5)
    canny = cv2.Canny(frame, 100, 200)

    cv2.imshow('webcam', canny)

    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

### image found edge

```python
img = cv2.imread('me1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

corrner = cv2.goodFeaturesToTrack(gray, 200, 0.1, 10)
corners = np.int0(corrner)

for corner in corners:
    x,y = corner.ravel()
    cv2.circle(img, (x,y), 3, (0,0,255), 1)

cv2.imshow('test', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
